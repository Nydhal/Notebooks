```
import requests as rq

class Tweet(object):
    def __init__(self, s, embed_str=False):
        if not embed_str:
            #Twitter oEmbed API https://dev.twitter.com/web/embedded-tweets
            api = '/oembed?url={}&hide_thread=true'.format(s)
            self.text = rq.get('https://publish.twitter.com'+api).json()["html"]
        else:
            self.text = s
    def _repr_html_(self):
        return self.text

Tweet("https://twitter.com/nntaleb/status/1080208165256523776")
```

```
from numpy import linspace,array,random,corrcoef
from matplotlib.pyplot import plot,title,show

n = 1000
for k in linspace(0,1,11):    
  X1 = random.normal(0,1,n)
  X2 = array([(X1[i]<0)*X1[i] + random.normal(0,1)*k for i in range(0,n)])  
  plot(X1,X2,'r.')
  title('Correlation: %s \n Gaussian Noise: %s'%(corrcoef(X1,X2)[0,1],k))
  show()
```
