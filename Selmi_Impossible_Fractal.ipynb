Render the Selmi Triangle using a 1-dimensional 4-state 6-cell-neighborhood cellular automaton.

```
#create selmi rule_1 dictionary by scanning temp string
temp='wwwwwwwrrgwwwwwwwwwwwwwrrgggwwwwwwwwwwwrrggbggwwwwwwwwwrrggwbbggwwwwwwwrr\
rrrrrbbggwwwwwrrgbbbbbrrgggwwwrrggbggwrrggbggwrrggwbbggrggwbbggrrrrrrbbgrrrrrbb\
ggbbbbbbbbbbbbbrrggbggwwwwwwwwwrrgg'
rule_1 = {''.join(temp[i:i+7]):temp[i+20] for i in range(len(temp)-20)}

#define first iteration row
seed = 21*'w'+'rrg'
curr = 21*'w'+1*seed+70*'w'

gens = 34  # number of generations

def draw(rule,curr):
  #main print/update loop
  for i in range(gens):
      print("%3i: %s"%(i,curr.replace('w',' ')\
                             .replace('r','\x1b[31m█\x1b[0m')\
                             .replace('g','\x1b[32m█')\
                             .replace('b','\x1b[34m█') ))
     #curr = curr[:-3]+curr+curr[:3])
      curr = 'www' + curr + 'www'
      curr = ''.join(rule[curr[i:i+7]] for i in range(len(curr)-6))

draw(rule_1,curr)
```

Introducing Random rules.

```
#create random rule_0
import itertools,random
rule_0 = {''.join(i):random.choice('wrgb') for i in\
 itertools.product('wrgb', repeat=7)}

#overwrite rule_0 with rule_1
rule = {**rule_0,**rule_1}

draw(rule,curr[20:-20]+seed[21:])
```
